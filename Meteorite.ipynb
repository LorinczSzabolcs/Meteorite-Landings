```python
csv_path='D:/Learning/DataScience/PlusCourses/DataVisualization/Meteorite/Meteorite_Landings.csv' 
```

```python
df=pd.read_csv(csv_path)
```

```python
df=df.drop(u'nametype', axis=1)
```

```python
df=df.drop(u'id', axis=1)
```

```python
df["index"] = [ i for i in range(len(df["name"]))]
```

```python
df.set_index(u'name')
```

```python
df2=df[:]
df2=df2.fillna(0)
```

```python
df2.rename(columns={'mass (g)': 'mass'}, inplace = True)
```

```python
gf=[]
for i in range(len(df2.index)):
    seged={}
    for j in range(len(df2.loc[i])):
        seged[df2.columns[j]]=str(df2.loc[i][j])
    gf.append(seged)
```

```python
import json
```

```python
file('D:/Learning/DataScience/PlusCourses/DataVisualization/Meteorite/Meteorite_Landings.json','w').write(json.dumps(gf))
```
